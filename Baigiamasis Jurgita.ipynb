# Istoriniai faktoriai

Šiandien nafta ir dujos yra multimilijoninė pramonė ir daugeliui kyla klausimas kaip ir kodėl ji prasidėjo, ir kodėl per paskutinius 10 metų atsirado žaliasis perversmas?


Tūkstančius metų prieš šiuolaikinę erą, kur ankstyviausi žinomi naftos gręžiniai buvo išgręžti Kinijoje 347 m., nafta ir dujos buvo naudojamos lempose ir kaip statybinė medžiaga.
Tikrasis perversmas įvyko, kai 1859 m. Edwinas Drake'as išgręžė pirmąjį modernų naftos gręžinį Titusvilyje, Pensilvanijoje. Energetikos Aukso amžius, prasidėjęs naftos atradimu, įtakojo daug naujų atradimų ir judėjimų pirmyn.



Naftos vartojimas 20-ame amžiuje pakeitė visuomenę dėl atsiradusių naujų produktų (kosmetika, asfaltas, valymo priemonės, medicinos preparatai, plastikas ir t.t.), tačiau tai pakeitė visuomenę ir dėl atsiradusio susidomėjimo ekologija ir naftos padariniais. Nenustojamos avarijos ir naftos išpylimai platformose bei klimato kaitos akivaizdus poveikis ozono sluoksniui bei atmosferos temperatūrai paskatino visuomenę susidomėti atsinaujinančia energija ir jos privalumais.

Dėl lėto valstybių reagavimo klimato kaitos klausimais, 2015 metais vykusioje Paryžiaus klimato kaitos konferencijoje su Greta Thunberg priešakyje, 196 pasaulio valstybės pasirašė paktą - Paryžiaus susitarimą. Šis susitarimas įpareigoja pasirašiusias valstybes mažinti CO2 išmetimą ir temperatūros didėjimą. Pagrindinis būdas tai padaryti - skatinti naftos produktų atsisakymą ir žalios energijos vartojimą.

```python
import pandas as pd
Contri=pd.read_csv("renewablesTotalPowerGeneration.csv", sep=",")
Contri.rename (columns= {'Contribution (TWh)': 'Contribution'}, inplace=True)
Contri
```

```python
Contri.plot(x='Mode of Generation')
plt.show()
```

```python
contribution = Contri['Contribution']
bigContri = 'Didelė energijos gamyba'
smallContri = 'Maža energijos gamyba'
cnt = list()
for val in contribution:
    if val > 2000:
        cnt.append(bigContri)
    else: 
        cnt.append(smallContri)
Contri['Gamyba'] = cnt
Contri
```
